```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import QFT
from qiskit import Aer, execute
from qiskit.visualization import plot_histogram
```

```python
n=5
qc = QuantumCircuit(n)
qc.ccx(0,1,3)
qc.ccx(2,3,4)
qc.ccx(0,1,3)
qc.draw('mpl')
```

```python
n=4
def phase_oracle():
    qc = QuantumCircuit(5)
    qc.ccx(0,1,3)
    qc.ccx(2,3,4)
    qc.ccx(0,1,3)
    
    return qc
```

```python
n=5
qc = QuantumCircuit(n)
for i in range(n-2):
    qc.x(i)
qc.ccx(0,1,3)
qc.ccx(2,3,4)
qc.ccx(0,1,3)
for i in range(n-2):
    qc.x(i)
qc.draw('mpl')
```

```python
def diffuser(n, name='V'):
    qc = QuantumCircuit(n, name=name)
    
    for qb in range(n-2): #first layer of Hadamards in diffuser
        qc.h(qb)
   
    for i in range(n-2):
        qc.x(i)
    qc.ccx(0,1,3)
    qc.ccx(2,3,4)
    qc.ccx(0,1,3)
    for i in range(n-2):
        qc.x(i)
    
    for qb in range(n-2): #second layer of Hadamards in diffuser
        qc.h(qb)
        
    return qc
```

```python
n=5
def Grover(n):
    gr = QuantumCircuit(n)
    gr.append(phase_oracle(n), range(n))
    gr.append(diffuser(n), range(n))
    return gr
```

```python
U_gate = phase_oracle().to_gate()
U_gate.label = "Gr Phase"
Con_U =U_gate.control()
```

```python
V_gate = diffuser(n).to_gate()
V_gate.label = "Diffuser"
Con_V =V_gate.control()
```

```python
n=5
c=4
qc=QuantumCircuit(n+c,c)

for qubit in range(n+c-2):
    qc.h(qubit)
# put ancilla in state |->
qc.x(n+c-1)
qc.h(n+c-1)
    
for x in range(c):
        exponent = 2**(c-x-1)
        for j in range(exponent):
            qc.append(Con_U,[x] + list(range(c, n+c)))
            qc.append(Con_V,[x] + list(range(c, n+c)))

qc.append(QFT(c).inverse(), range(c))       

for i in range(c):
    qc.measure(i,i)

qc.draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
counts = execute(qc, backend=simulator, shots=10000).result().get_counts()
plot_histogram(counts)
```

```python
import operator
highest_probability_string = max(counts.items(), key=operator.itemgetter(1))[0][::-1]
highest_probability_string
measured_int=int(highest_probability_string,2)
```

```python
import math as mt
N=2**(n-2)
theta=2*mt.pi*measured_int/(2**c)
M=N*(mt.sin(theta/2))**2
print('M = ',M)
print('N = ',N)
print('Number of solutions = ',N-M)
```
